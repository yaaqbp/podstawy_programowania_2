Chcemy napisać funkcję zwracającą ciąg arytmetyczny (skończony fragment). Funkcja powinna przyjmować trzy parametry:
* pierwszy wyraz,
* różnica ciągu,
* liczba wyrazów.

Ciąg powinien zostać zwrócony w postaci listy liczb.

### 2.1

Zaproponuj testy dla tej funkcji.

### 2.2

Napisz funkcję i sprawdź, że przechodzi Twoje testy.

````python
def aryt(a1, r, n):
    try:
        a1 =float(a1)
        r = float(r)
        n = int(n)
        if n == 0:
            return []
        lst = [a1]
        for i in range(n-1):
            lst.append(a1+r*(i+1))
        return lst
    except ValueError:
        return "złe dane"
````

````python
assert aryt(0, 0, 0) == [], 'Zero wyrazów to ciag pusty'
assert aryt(1,1,1) == [1]
assert aryt(1,1,10) == list(range(1,11))
assert aryt(2.5,.5,3) == [2.5,3,3.5]
assert aryt(0,-0.5, 4) == [0,-0.5,-1,-1.5]
````

### 4.2


Napisz dwuparametrową funkcję `szyfruj_1(tekst, klucz)`. Argumenty `klucz` i `tekst` są łańcuchami. Funkcja zwraca łańcuch będący zaszyfrowanym tekstem `tekst` według klucza `klucz`. Funkcja powinna mieć argument domyślny `klucz` ustawiony na wartość `'malinowebuty'`.

#### Testy

````python
# W słowniku szyfry_test klucz to tekst do zaszyfrowania, 
# wartość to klucz po zaszyfrowaniu. 
szyfry_test = {'a': 'm', 'm': 'a', 
               'x': 'x', 'A': 'A', 
               'malinowebuty': 'amilonewubyt', 
               'ziemia jest płaska': 'zlwalm jwsy płmskm'}

for jawny, tajny in szyfry_test.items():
    assert szyfruj_1(jawny) == tajny, 'Błąd, powinno być: {} ==> {}'.format(jawny, tajny)

# Krotki (klucz, tekst_jawny, tekst_tajny)
test_klucz_jawny_tajny = [('ab', 'abc', 'bac'),
                          ('XYUV', 'xVyU', 'xUyV'),
                          ('gaderypoluki', 'gaderypoluki', 'agedyropulik')]

for klucz, jawny, tajny in test_klucz_jawny_tajny:
    assert szyfruj_1(jawny, klucz) == tajny, 'Klucz {} zamienia {} na {}'.format(klucz, jawny, tajny)
````

### 4.3


Klucz uznajemy za prawidłowy, gdy jest różnowartościowy i ma parzystą długość. Napisz funkcję  `szyfruj_2(tekst, klucz)` rozszerzającą `szyfruj_1()` o test czy klucz jest poprawny. W przypadku wykrycia nieprawidłowego klucza funkcja `szyfruj_2()` ma wywoływać wyjątek `ValueError` z komunikatem: `nieprawidłowy klucz!`.

#### Testy

````python
# W słowniku szyfry_test klucz to tekst do zaszyfrowania, 
# wartość to klucz po zaszyfrowaniu. 
szyfry_test = {'a': 'm', 'm': 'a', 
               'x': 'x', 'A': 'A', 
               'malinowebuty': 'amilonewubyt', 
               'ziemia jest płaska': 'zlwalm jwsy płmskm'}

for jawny, tajny in szyfry_test.items():
    assert szyfruj_2(jawny) == tajny, 'Błąd, powinno być: {} ==> {}'.format(jawny, tajny)
    
# Nieprawidłowe klucze do testowania wyjątków.
nieprawidłowe_klucze = ['x', 'xyz', 'xyzx']

for klucz in nieprawidłowe_klucze:
    try:
        szyfruj_2('abc', klucz)
    except ValueError as err:
        assert str(err) == 'nieprawidłowy klucz!', 'Niewłaściwy komunikat.'
    except Exception as err:
        raise AssertionError('Nieprawidłowy wyjątek {!r}'.format(err))
    else:
        raise AssertionError("Brak wyjątku: '{}' nie jest prawidłowym kluczem!".format(klucz))

# Krotki (klucz, tekst_jawny, tekst_tajny)
test_klucz_jawny_tajny = [('ab', 'abc', 'bac'),
                          ('XYUV', 'xVyU', 'xUyV'),
                          ('gaderypoluki', 'gaderypoluki', 'agedyropulik')]

for klucz, jawny, tajny in test_klucz_jawny_tajny:
    assert szyfruj_2(jawny, klucz) == tajny, 'Klucz {} zamienia {} na {}'.format(klucz, jawny, tajny)
````
